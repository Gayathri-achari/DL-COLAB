Inspect the data of the weather dataset

```python
import pandas as pd

# Load the weather dataset
weather_data = pd.read_csv('seattle-weather (1).csv')

# Inspect the data
print(weather_data.head())
print(weather_data.info())
print(weather_data.describe())
```

Parsing the data

```python
# Parse the Date column
weather_data['date'] = pd.to_datetime(weather_data['date'])
```

Plotting the temperature timeseries

```python
import matplotlib.pyplot as plt

# Plot the temperature timeseries
plt.figure(figsize=(10, 6))
plt.plot(weather_data['date'], weather_data['temp_max'])
plt.xlabel('date')
plt.ylabel('Temperature (Â°C)')
plt.title('Temperature Timeseries')
plt.show()
```

Plotting the first 10 days of the temperature timeseries

```python
# Plot the first 10 days of the temperature timeseries
plt.figure(figsize=(10, 6))
plt.plot(weather_data['date'][:10], weather_data['temp_max'][:10])
plt.xlabel('date')
plt.ylabel('Temperature (Â°C)')
plt.title('First 10 Days of Temperature Timeseries')
plt.show()
```
